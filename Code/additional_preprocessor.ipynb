# UPdate steps - convert to list in smaller_csv.

```python
csv_path = os.path.join(base_path, dataset_dir)

big_csv_path, small_csv_path = csv_list

small_csv = pd.read_csv(os.path.join(csv_path, small_csv_path))
```

```python
small_csv.head()
```

```python
# Find all null.

small_csv.isnull().sum()
```

```python
small_csv.loc[small_csv['Instructions'].isnull()]
```

```python
# Delete all rows where Instructions and Title are NULL.
small_csv.drop(small_csv.loc[small_csv['Instructions'].isnull() & small_csv['Title'].isnull()].index, inplace=True)

```

```python
small_csv.loc[small_csv['Instructions'].isnull()]
```

```python
def update_instructions(column_data):
    """
        This function splits the input str column to list.
    """
    if isinstance(column_data, float):
        return []
    column_data = column_data.split('.')

    return column_data


```

```python
small_csv['split_steps'] = small_csv['Instructions'].apply(update_instructions)
```

```python
small_csv.to_csv(os.path.join(csv_path, small_csv_path), index = False)
```
