## Remove Fractions


```python
# Deleting fractions

# Just delete the fraction symbols
fraction_removal_table = str.maketrans('', '', '¼½¾⅐⅑⅒⅓⅔⅕⅖⅗⅘⅙⅚⅛⅜⅝⅞')
```

```python
small_csv['Ingredients'] = small_csv['Ingredients'].str.translate(fraction_removal_table)
```

```python
small_csv
```

```python
def remove_punctuation(column_data):
    column_data = "".join([word for word in column_data if word not in string.punctuation])
    column_data = column_data.lower()
    column_data = re.sub('[0-9]+', '', column_data) # remove_numbers
    #tokenize
    column_data = re.split(r"\W+", column_data)
    
    return column_data
```

```python
small_csv['Ingredients_tokenized'] = small_csv['Ingredients'].apply(remove_punctuation)
```

```python
# Combination of lists:


word_quantities = [
    "one", "two", "three", "four", "five", "six", "seven", "eight", "nine", "ten",
    "dozen", "half", "quarter"
]
optional_keywords = [
    "optional", "to taste", "as needed", "for garnish", "for frying", 
    "for serving", "for decoration", "plus more", "divided", "or more", "or to taste"
]
packaging_types = [
    "can", "cans", "jar", "jars", "bottle", "bottles", "package", "packages", 
    "bag", "bags", "box", "boxes", "carton", "cartons", "tub", "tubs", "packet", "packets"
]
state_descriptors = [
    "cold", "hot", "warm", "room temperature", "frozen", "fresh", "dried", "raw",
    "cooked", "uncooked", "boiled", "baked", "steamed", "chilled", "cooled", "thawed"
]

preparation_methods = [
    "chopped", "sliced", "diced", "minced", "grated", "shredded", "peeled", "crushed",
    "ground", "mashed", "pureed", "beaten", "whisked", "toasted", "roasted", "baked", "steamed",
    "boiled", "drained", "melted", "blended", "cubed", "rinsed", "halved", "quartered", 
    "julienned", "zested", "frozen", "thawed", "trimmed", "seeded", "cooked", "uncooked", 
    "deveined", "shelled", "soaked", "dried", "pitted", "smashed", "sautéed", "cut", "inch", "dice", "roll"
]
portion_sizes = [
    # Fractions & Quantifiers
    "whole", "half", "quarter", "third", "fourth", "fifth", "eighth", 
    "1/2", "1/3", "1/4", "2/3", "3/4", "3/5", "1/8", "2/5", "3/8", "5/8", "7/8",
    
    # Teaspoon/Tablespoon abbreviations
    "tsp", "teaspoon", "teaspoons",
    "tbsp", "tablespoon", "tablespoons",
    
    # Volume-based
    "cup", "cups", "pint", "pints", "quart", "quarts", "gallon", "gallons", "ml", "milliliter", "milliliters", "l", "liter", "liters",

    # Weight-based
    "oz", "ounce", "ounces", "gram", "grams", "kg", "kilogram", "kilograms", "pound", "pounds", "lb", "lbs",

    # Unit-based
    "piece", "pieces", "slice", "slices", "clove", "cloves", "stick", "sticks", "can", "cans", "bag", "bags", "bottle", "bottles",

    # Descriptors (commonly used with portions)
    "small", "medium", "large", "extra large", "extra-small", "extra-large"
]

ingredient_colors = [
    "white", "black", "red", "green", "blue", "yellow", "brown", "orange", "purple", "pink",
    "gray", "grey", "gold", "silver", "tan", "beige", "ivory", "peach", "maroon", "olive",
    "navy", "turquoise", "teal", "charcoal", "lavender", "mustard", "copper", "bronze",
    "burgundy", "amber", "aqua", "coral", "fuchsia", "mint", "salmon", "cream", "chocolate",

    # Compound shades
    "light brown", "dark brown", "light green", "dark green", "light blue", "dark blue",
    "dark red", "light red", "dark yellow", "light yellow", "pale yellow", "deep red",
    "off white", "dark chocolate", "bright red", "golden brown", "reddish", "bluish",
    "whitish", "blackish", "greenish"
]
```

```python
master_list = []
master_list.extend(word_quantities)
master_list.extend(optional_keywords)
master_list.extend(packaging_types)
master_list.extend(state_descriptors)
master_list.extend(preparation_methods)
master_list.extend(portion_sizes)
master_list.extend(ingredient_colors)
```

```python
len(master_list)
```

## Filter on Master list

```python
def filter_master_list(column_data):
    feature = [ing for ing in column_data if ing not in master_list and ing]
    return feature
```

```python
small_csv['Ingredients_tokenized'] = small_csv['Ingredients_tokenized'].apply(filter_master_list)
```

```python
small_csv
```

```python

stopwords = nltk.corpus.stopwords.words('english')
```

```python
# Remove stopwords
def remove_stopwords(column_data):
    feature = [ing for ing in column_data if ing not in stopwords and ing]
    return feature

```

```python
small_csv['Ingredients_tokenized'] = small_csv['Ingredients_tokenized'].apply(remove_stopwords)

small_csv.tail()
```

```python
# LEMMATIZE MASTER LIST
my_lemmatizer = nltk.WordNetLemmatizer()
lemmatized_master_list = [my_lemmatizer.lemmatize(entity) for entity in master_list]
```

```python
lemmatized_master_list[:20]
```

```python
# Remove stopwords from lemmatized_master_list
lemmatized_master_list = [ing for ing in lemmatized_master_list if ing not in stopwords and ing]
lemmatized_master_list[:20]
```

```python
for idx, phrase in enumerate(lemmatized_master_list):
    sub_split = phrase.split()  # Split the phrase into words
    # Remove stopwords using list comprehension
    if len(sub_split) > 1:
        filtered_words = [word for word in sub_split if word not in stopwords]
        # Join the remaining words back into a phrase and strip any leading/trailing spaces
        processed_phrase = ' '.join(filtered_words).strip()
        lemmatized_master_list[idx] = processed_phrase  # Add the processed phrase to the list
```

```python
lemmatized_master_list[:20]
```

# # Remove Lemmatized Master List words

```python
# Remove lemmatized words
def remove_lemmatized_words(column_data):
    feature = [ing for ing in column_data if ing not in lemmatized_master_list and ing]
    return feature

```

```python
small_csv['Ingredients_tokenized'] = small_csv['Ingredients_tokenized'].apply(remove_lemmatized_words)

small_csv.tail()
```
