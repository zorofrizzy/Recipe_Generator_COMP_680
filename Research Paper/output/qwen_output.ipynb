```python
import matplotlib.pyplot as plt

severity_counts = df_risks['Severity'].value_counts()

plt.figure(figsize=(7, 7))
severity_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff4d4d', '#ffa500', '#ffd700'])
plt.title('Distribution of Risks by Severity - Qwen')
plt.ylabel('')
plt.show()
```

```python
phase_counts = df_risks['SDLC_Phase'].value_counts()

plt.figure(figsize=(8, 5))
phase_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Risks by SDLC Phase - Qwen')
plt.xlabel('SDLC Phase')
plt.ylabel('Number of Risks')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

heatmap_data = df_risks.pivot_table(index='Likelihood', columns='Impact', aggfunc='size', fill_value=0)

plt.figure(figsize=(7, 5))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', cbar=False, fmt='g')
plt.title('Risk Heatmap: Likelihood vs Impact - Qwen')
plt.xlabel('Impact')
plt.ylabel('Likelihood')
plt.tight_layout()
plt.show()
```

```python
grouped = df_risks.groupby(['SDLC_Phase', 'Risk_Type']).size().unstack(fill_value=0)

grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=sns.color_palette("Set2"))
plt.title('Risk Types Across SDLC Phases - Qwen')
plt.xlabel('SDLC Phase')
plt.ylabel('Number of Risks')
plt.xticks(rotation=45)
plt.legend(title='Risk Type')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
